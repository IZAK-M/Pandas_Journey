#### Groupby et agrégations

```python
df_ventes = pd.read_csv('data/ventes.csv')

df_ventes.head(5)
```

```python
# Somme par groupe 
par_region = df_ventes.groupby('region')['ventes'].sum()
par_region
```

```python
# Plusieurs agrégations
stats = df_ventes.groupby('region').agg( 
    total_ventes =('ventes', 'sum'),
    moyenne    = ('ventes', 'mean'),
    nb_lignes  = ('ventes', 'count'),
    max_ventes = ('ventes', 'max')
    )

print(stats.reset_index())
```
