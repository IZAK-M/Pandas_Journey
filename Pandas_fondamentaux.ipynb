#### Groupby et agrégations

```python
df_ventes = pd.read_csv('data/ventes.csv')

df_ventes.head(5)
```

```python
# Somme par groupe 
par_region = df_ventes.groupby('region')['ventes'].sum()
par_region
```

```python
# Plusieurs agrégations
stats = df_ventes.groupby('region').agg( 
    total_ventes =('ventes', 'sum'),
    moyenne    = ('ventes', 'mean'),
    nb_lignes  = ('ventes', 'count'),
    max_ventes = ('ventes', 'max')
    )

print(stats.reset_index())
```

#### Merge / join entre deux tables

```python
clients = pd.read_csv('data/clients.csv')
commandes = pd.read_csv('data/commandes.csv')
```

```python
# Inner join
df_inner = pd.merge(clients, commandes, on='client_id', how='inner')
df_inner
```

```python
# Lest join
df_left = pd.merge(clients, commandes, on='client_id', how='left')
df_left
```

```python
# Vérification des lignes non matchées
non_matches = df_left[df_left['commande_id'].isnull()]
non_matches
```

```python
df_right = pd.merge(clients, commandes, on='client_id', how='right')
df_right
```

#### Nettoyer des données (valeurs nulles)

```python
df_data = pd.read_csv('data/data.csv')
```

```python
# Voir les nulls
df_data.isnull().sum()
```

```python
# Supprimer les lignes avec trop de nulls
df_data = df_data.dropna(thresh=5)
```

```python
# Remplir les nulls numériques par la médiane
df_data['age']= df_data['age'].fillna(df_data['age'].median())
df_data['salaire']= df_data['salaire'].fillna(df_data['salaire'].mean())
```

```python
# Remplir les nulls texte par 'Inconnu'
df_data['ville']= df_data['ville'].fillna('Inconnu')
df_data['statut']= df_data['statut'].fillna('Inconnu')
```

```python
# Vérification
df_data.isnull().sum()
```

#### Création et transformation des colonnes

```python
df_ventes_v2 = pd.read_csv('data/ventes_v2.csv')
```

```python
# Nouvelle colonne calculée
df_ventes_v2['marge'] = df_ventes_v2['prix'] - df_ventes_v2['cout']
```

```python
# Colonne conditionnelle 
df_ventes_v2['categorie']= df_ventes_v2['marge'].apply(
    lambda x: 'bon' if x > 100 else ('moyen' if x >= 50 else 'faible')
)
```

```python
# Transformation de texte
df_ventes_v2['region']= df_ventes_v2['region'].str.upper().str.strip()
```

```python
# Convertir un type
df_ventes_v2['date'] = pd.to_datetime(df_ventes_v2['date'])
```

#### Sauvegarder et exporter des données

```python
df_data = pd.read_csv('data/data.csv')
```

```python
# Nettoyage minimal
df_data = df_data.dropna()
df_data.columns = [c.lower().strip() for c in df_data.columns]
```

```python
# Export CSV
df_data.to_csv('data/export/data_propre.csv', index=False)

# Export Excel
df_data.to_excel('data/export/data_propre.xlsx', index=False, sheet_name='Data')

# Export partiel (filtre avant export)
df_data[df_data['statut'] == 'actif'].to_csv('data/export/actif.csv', index=False)

#Export JSON
df_data.to_json('data/export/data_propre.json', orient='records', indent=2)
print("Exports terminés")
```
